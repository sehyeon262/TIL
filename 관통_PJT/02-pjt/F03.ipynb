```python
import numpy as np
import pandas as pd
```

```python
def file_open_by_numpy():
    np_arr = np.loadtxt('NFLX.CSV', delimiter=",", encoding='cp949', dtype=str) # 텍스트 파일을 읽어와 Numpy 배열로 변환
    return np_arr

df = pd.read_csv('NFLX.csv', encoding = 'cp949', usecols = ['Date', 'Open', 'High', 'Low', 'Close'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Data 값을 날짜 객체로 변환 
```

```python
df_filter = df.loc[df['Date'] >= '2021-01-01', ['Date', 'Open', 'High', 'Low', 'Close']]
```

```python
max_close = df_filter['Close'].max()
min_close = df_filter['Close'].min()
print("죄고 종가: ", max_close)
print("최소 종가: ", min_close)
```
