```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 파일 열기

```python
def file_open_by_numpy():
    np_arr = np.loadtxt('NFLX.CSV', delimiter=",", encoding='cp949', dtype=str) # 텍스트 파일을 읽어와 Numpy 배열로 변환
    return np_arr

df = pd.read_csv('NFLX.csv', encoding = 'cp949', usecols = ['Date', 'Open', 'High', 'Low', 'Close'])
```

### 데이터 형 변환

```python
df['Date'] = pd.to_datetime(df['Date'])  # Data 값을 날짜 객체로 변환 
```

### 2021년 이후 데이터 필터링

```python
df_filter = df.loc[df['Date'] >= '2021-01-01', ['Date', 'Open', 'High', 'Low', 'Close']]
```

### 그래프 시각화

```python
plt.figure(figsize=(6, 5))
plt.plot(df_filter['Date'], df_filter['Close'], linewidth = 1.5)
plt.title('NFLX Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45) # x축 글씨 45도 회전 
plt.tight_layout()
plt.show()
```
